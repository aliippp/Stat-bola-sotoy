```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Arc
from mplsoccer import Pitch, VerticalPitch
%matplotlib inline
```

```
data = pd.read_csv("events_3794686_pass.csv")
data.head()
```

```
laporte = data[data["player"] == "Aymeric Laporte"]
laporte.head()
```

```
fig, ax = plt.subplots()
fig.set_size_inches(7, 5)

for i in range(len(data)):
    plt.plot([int(data["start_location_x"][i]),int(data["end_location_x"][i])],
             [int(data["start_location_y"][i]),int(data["end_location_y"][i])], 
             color="blue")
    
plt.show()
```

```
spain = data[data['possession_team'] == 'Spain'].reset_index().drop('level_0', axis=1)
spain
```

```
pitch = Pitch()
# specifying figure size (width, height)
fig, ax = pitch.draw(figsize=(8, 4))

for i in range(len(spain)):
    plt.plot([int(spain["start_location_x"][i]),int(spain["end_location_x"][i])],[int(spain["start_location_y"][i]),int(spain["end_location_y"][i])], color="#aa151b")

#tile
plt.title('Passmap Spain vs Croatia Euro 2021 (Spain)')
#Display Pitch
plt.show()
```

```
#filling na in outcome value
spain.outcome.fillna('Complete', inplace=True)
```

```
#creating specific variable
laporte = spain[spain['player'] == 'Aymeric Laporte'].reset_index().drop('level_0', axis=1)
laporte.head()
```

```
#setting pitch
fig ,ax = plt.subplots(figsize=(13.5,8))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')

#this is how we create the pitch
pitch = Pitch(pitch_type='statsbomb', orientation='horizontal',
              pitch_color='#22312b', line_color='#c7d5cc', figsize=(13, 8),
              constrained_layout=False, tight_layout=True)

#Draw the pitch on the ax figure as well as invert the axis for this specific pitch
pitch.draw(ax=ax)
plt.gca().invert_yaxis()
for i in range(len(laporte)):
    if laporte['outcome'][i] == 'Complete':
        plt.plot([int(laporte["start_location_x"][i]),int(laporte["end_location_x"][i])],[int(laporte["start_location_y"][i]),int(laporte["end_location_y"][i])], color="#00FF00")
        plt.plot(int(laporte["start_location_x"][i]),int(laporte["start_location_y"][i]),"o", color="green")
    elif laporte['outcome'][i] == 'Incomplete':
         plt.plot([int(laporte["start_location_x"][i]),int(laporte["end_location_x"][i])],[int(laporte["start_location_y"][i]),int(laporte["end_location_y"][i])], color="#aa151b")
         plt.plot(int(laporte["start_location_x"][i]),int(laporte["start_location_y"][i]),"o", color="red")

#title
plt.title('Laporte vs Croatia Euro 2021', color='white', size=20)
fig.tight_layout()
plt.savefig("laporte_pass_cro.png")
#Display Pitch
plt.show()
```
